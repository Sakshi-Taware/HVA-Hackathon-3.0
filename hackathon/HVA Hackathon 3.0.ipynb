1. Factors Influencing Happiness Scores: Analyze the impact of different factors such as
GDP per capita, social support, life expectancy, freedom to make life choices, generosity,
and perceptions of corruption on overall happiness scores across different countries.


```python
# Scatter plot for GDP per capita vs Happiness Score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='GDP per capita', y='Happiness Score')
plt.title('GDP per Capita vs Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()

```

```python
# Correlation matrix

correlation_matrix = data[['Happiness Score', 'GDP per capita', 'Social support', 'Life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Happiness Factors')
plt.show()
```

2. Regional Happiness Disparities: Explore regional disparities in happiness scores.
Investigate whether certain regions consistently rank higher or lower in happiness
compared to others. Identify any commonalities or differences in the contributing
factors among these regions


```python
# Group by countries  and calculate average happiness scores
region_happiness = data.groupby('Country')['Happiness Score'].mean().sort_values(ascending=False)

# Visualize regional disparities
plt.figure(figsize=(15, 6))
region_happiness.plot(kind='bar')
plt.title('Average Happiness Score by Region')
plt.xlabel('Region')
plt.ylabel('Average Happiness Score')
plt.xticks(rotation=45)
plt.show()
```

3. Trends Over Time: Examine how happiness scores have changed over time for different
countries. Identify any significant trends or patterns in happiness levels and investigate
potential factors contributing to these changes.


```python
# Plot trend of happiness scores over years
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='Happiness Score', hue='Country')
plt.title('Trend of Happiness Scores Over Time')
plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

4. Correlation Analysis: Conduct correlation analysis to determine the relationships
between happiness scores and other variables in the dataset. For example, analyze the
correlation between GDP per capita and happiness scores, social support and happiness
scores, etc.


```python
# Calculate correlation between GDP per capita and happiness scores
gdp_corr = data['GDP per capita'].corr(data['Happiness Score'])
print(f"Correlation between GDP per capita and Happiness Score: {gdp_corr}")
```
